```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df=pd.read_excel(r"F:\python code\pythonproject\nycovid19.xlsx")
df.head
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR


x=df.index.values.tolist()
y=df['positive']
x = np.array(x) 
x=x.reshape(-1,1)
model=SVR(shrinking=True, kernel='poly',degree=5)




x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.25,shuffle=False)
```

```python
model.fit(x,y)
y_predict=model.predict(x_test)
print(r2_score(y_test,y_predict))
print(mean_squared_error(y_test,y_predict))
```

```python
plt.plot(x_test,y_predict)
```

```python
test_arr=np.arange(180,300).reshape(-1,1)
pred_arr=model.predict(test_arr)
plt.plot(test_arr,pred_arr,label="Prediction")
plt.plot(df.index.values.tolist(),y,'r', label="Current data")
plt.legend(loc="upper left")
plt.show()
```
